```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate dummy data for experience (years) and salary (in USD)
experience = np.random.randint(1, 21, size=1000)  # Experience from 1 to 20 years
slope = 5000  # A slope of 5000 USD increase in salary per year of experience
intercept = 30000  # Starting salary when experience is 0
noise = np.random.normal(0, 10000, size=1000)  # Random noise to add to the salary

# Generate salary data (linear relationship with some noise)
salary = slope * experience + intercept + noise

# Create a pandas DataFrame
data = pd.DataFrame({
    'Experience (Years)': experience,
    'Salary (USD)': salary
})

# Show the first few rows of the generated data
data.head()

```

```python
x=data.iloc[:,0:1].values
y=data.iloc[:,1].values
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
```

```python
print(regressor.coef_[0])

print(regressor.intercept_)
```

```python
# Predict y values for the entire x range for plotting the regression line
y_pred = regressor.predict(x_test)

# Plotting the data points and the regression line
plt.scatter(x_test, y_test, color='blue', label='Actual Data Points')  # Actual test data points
plt.plot(x_test, y_pred, color='red', label='Regression Line')  # Predicted regression line
plt.title('Linear Regression: Actual Data vs. Predicted Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary (USD)')
plt.legend()
plt.show()
```
